###let replace the missing

```python
# for rating (using default)

# note that to use mode as auto refill :
# mode_rating = df['rating'].mode()[0]  # Get the most common rating
# df['rating'].fillna(mode_rating, inplace=True)

df['rating'].fillna('Not Rated', inplace=True)
```

```python
# for release (using default)
# note that to use mode as auto refill :
#mode_released = df['released'].mode()[0]  # Get the most common release date
#df['released'].fillna(mode_released, inplace=True)

df['released'].fillna('Unknown Release Date', inplace=True)
```

```python
# for score (using mean)
# note that to use median as auto refill especially when there are outliers:
#median_score = df['score'].median()  # Calculate the median
#df['score'].fillna(median_score, inplace=True)

mean_score = df['score'].mean()  # Calculate the mean
df['score'].fillna(mean_score, inplace=True)
```

```python
# for vote (using median because the mean is too far away feom the median, hence their might be an outliers)
#to use mean as auto refill :
#mean_votes = df['votes'].mean()  # Calculate the mean
#df['votes'].fillna(mean_votes, inplace=True)

median_votes = df['votes'].median()  # Calculate the median
df['votes'].fillna(median_votes, inplace=True)
```

```python
# For writer column (using mode)
mode_writer = df['writer'].mode()[0]  # Get the mode
df['writer'].fillna(mode_writer, inplace=True)

# For star column (using mode)
mode_star = df['star'].mode()[0]  # Get the mode
df['star'].fillna(mode_star, inplace=True)

# For country column (using mode)
mode_country = df['country'].mode()[0]  # Get the mode
df['country'].fillna(mode_country, inplace=True)
```

```python
# For budget column, using mean
mean_budget = df['budget'].mean()  # Calculate the mean
df['budget'].fillna(mean_budget, inplace=True)

# For gross column, using median (because there seem to be an outlier)
median_gross = df['gross'].median()  # Calculate the median
df['gross'].fillna(median_gross, inplace=True)
```

```python
# For company column, using mode
mode_company = df['company'].mode()[0]  # Get the mode
df['company'].fillna(mode_company, inplace=True)

# For runtime column, using mean
mean_runtime = df['runtime'].mean()  # Calculate the mean
df['runtime'].fillna(mean_runtime, inplace=True)
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```
